## while loop

Python 编程中 while 语句用于循环执行程序，即在某条件下，循环执行某段程序，以处理需要重复处理的相同任务。



这是一个演示：

```python
i = 0

while i < 10:
    print(i)
    i += 1
```

在这一段代码中，i的值在循环中发生了改变，最终导致while循环之后的条件判断结果为False，从而结束while循环。

或者我们可以使用break关键字：

```python
i = 0
cout = 0

while i < 100:
    print(cout)
    cout += 1
    if cout == 23:
        break
```

```python
L = []

while len(L) < 5:
    new_name = input("Please add a new name: ").strip().capitalize()
    L.append(new_name)

print("Sorry list is full")
print(L)
```

## for loop

Python for循环可以遍历任何序列的项目，如一个列表或者一个字符串。

```python
list_one = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
for num in list_one:
    print(num)
```

```python
students = {
    "male": ["Tom", "Charlie", "Harry", "Frank"],
    "female":["Sarah", "Huda", "Samantha", "Emily", "Elizabeth"]
    }

for key in students.keys():
    for name in students[key]:
        if "a" in name:
            print(name)

```

for可以和else进行结合：

```python
for num in range(10,20):  # 迭代 10 到 20 之间的数字
   for i in range(2,num): # 根据因子迭代
      if num%i == 0:      # 确定第一个因子
         j=num/i          # 计算第二个因子
         print ('%d 等于 %d * %d' % (num,i,j))
         break            # 跳出当前循环
   else:                  # 循环的 else 部分
      print ('%d 是一个质数' % num)
```

for … else 表示这样的意思，for 中的语句和普通的没有区别，else 中的语句会在循环正常执行完（即 for 不是通过 break 跳出而中断的）的情况下执行，while … else 也是一样。

重点：不推荐使用for-else和while-else语句。
