## 什么是变量

我们可以这样去看待变量：



这是一个小盒子，这个小盒子代表的是在内存中开辟的内存空间，这个内存空间有一个内存地址，而这个内存空间中会存储数据，因为我们人并不能够很好的记住内存空间的地址，所以我们会给这个内存空间一个名字，这个名字就是变量名。

当我们使用这个变量名的时候，编译器会自动去获取这个变量名所对应内存空间中的数据。

```python
number = 1  # 我们创建了一个变量，变量名为number，存储的数据为1
```

```python
number
```

Python语言会区分大小写：

```python
NUMBER
```

可以发现NUMBER和number并不会被认为是同一个变量。

我们可以使用type()来查看一个变量的类型：

```python
type(number)
```

```python
type('string')
```

```python
type(10.10)
```

## Updating our Script

我们来创建一个first_name变量：

```python
first_name = "John"
```

接着创建一个last_name变量：

```python
last_name = "Smith"
```

然后我们可以将这两个变量合并并且赋值给一个新的变量：

```python
full_name = first_name + last_name
```

```python
full_name
```

我们可以再修改一下格式：

```python
full_name = first_name + " " + last_name
```

```python
full_name
```

=或进行赋值，将=号的右边的值赋值给左边。
